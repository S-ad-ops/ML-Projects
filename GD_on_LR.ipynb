```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/input.csv.csv")
print(df)
```

```python
X = df["YearsExperience"].values
y = df["Salary"].values
```

```python
m = 0
c = 0
```

```python
learning_rate = 0.001
epochs = 1000
```

```python
def compute_cost(y, y_predicted):
    return np.mean((y - y_predicted) ** 2)
```

```python
n = len(X)
cost_hist = []
for epoch in range(epochs):
  y_predicted = m * X + c
  dm = (-2 / n) * np.sum(X * (y - y_predicted))
  dc = (-2 / n) * np.sum(y - y_predicted)
  m = m - learning_rate * dm
  c = c - learning_rate * dc
  cost = compute_cost(y, y_predicted)
  cost_hist.append(cost)
  if epoch % 100 == 0:
    print(f"Epoch {epoch}: Cost = {cost}")
```

```python
print("Slope:",m)
print("Intercept:",c)
```

```python
experience = 7
salary = m * experience + c
print(salary)
```

```python
plt.scatter(X,y)
plt.plot(X,m*X+c,color="red")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary vs Experience")
plt.show()
```

```python
plt.plot(cost_hist)
plt.xlabel("Epochs")
plt.ylabel("Cost")
plt.title("Cost vs Epochs")
plt.show()
```
